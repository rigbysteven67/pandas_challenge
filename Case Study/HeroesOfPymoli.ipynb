```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
num_of_players = purchase_data['SN'].nunique()
num_of_players_df = pd.DataFrame([{'Number of Players':num_of_players}])
num_of_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
num_of_unique_items = len(purchase_data["Item ID"].unique())
total_of_purchases = purchase_data["Price"].sum()
num_of_purchases = len(purchase_data['Price'])
avg_purchase_price = total_of_purchases/num_of_purchases
purchasing_df = pd.DataFrame({"Number of Unique Items": [num_of_unique_items],
                             "Average Price": [avg_purchase_price],
                             'Number of Purchases': [num_of_purchases],
                             'Total Revenue': [total_of_purchases]})
purchasing_df
```

## Gender Demographics

```python
removed_dupes_df = purchase_data.drop_duplicates(subset=['SN'])
male_count = len(removed_dupes_df.loc[removed_dupes_df['Gender'] == 'Male', :])
male_percent = (male_count / num_of_players) * 100
female_count = len(removed_dupes_df.loc[removed_dupes_df['Gender'] == 'Female', :])
female_percent = (female_count / num_of_players) * 100
other_count = len(removed_dupes_df.loc[removed_dupes_df['Gender'] == 'Other / Non-Disclosed', :])
other_percent = (other_count / num_of_players) * 100
gender_demographics_df = pd.DataFrame({'Gender': ['Male', 'Female', 'Other / Non-Discosed'],
                      'Total Count': [male_count, female_count, other_count],
                      'Percentage of Players': [male_percent, female_percent, other_percent]})
gender_demographics_df
```

## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
male_purchases = purchase_data.loc[purchase_data['Gender'] == 'Male', :]
male_purchases_count = len(male_purchases['Purchase ID'])
male_purchases_total = male_purchases['Price'].sum()
male_purchases_avg = male_purchases_total / male_purchases_count
male_purchases_avg_person = male_purchases_total / male_count
```

```python
female_purchases = purchase_data.loc[purchase_data['Gender'] == 'Female', :]
female_purchases_count = len(female_purchases['Purchase ID'])
female_purchases_total = female_purchases['Price'].sum()
female_purchases_avg = female_purchases_total / female_purchases_count
female_purchases_avg_person = female_purchases_total / female_count
```

```python
other_purchases = purchase_data.loc[purchase_data['Gender'] == 'Other / Non-Disclosed', :]
other_purchases_count = len(other_purchases['Purchase ID'])
other_purchases_total = other_purchases['Price'].sum()
other_purchases_avg = other_purchases_total / other_purchases_count
other_purchases_avg_person = other_purchases_total / other_count
```

```python
purchasing_analysis_gender_df = pd.DataFrame({"Gender": ['Male', 'Female', 'Other / Non-Discolsed'],
                                            'Purchase Count': [male_purchases_count, female_purchases_count, other_purchases_count],
                                              'Average Purchase Price': [male_purchases_avg, female_purchases_avg, other_purchases_avg],
                                              'Total Purchase Value': [male_purchases_total, female_purchases_total, other_purchases_total],
                                              'Avg Total Purchase per Person': [male_purchases_avg_person, female_purchases_avg_person, other_purchases_avg_person]})
purchasing_analysis_gender_df
                                              
```
